```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import pandas as pd
```

1.
1a.

```python
np.random.seed(150)
X1=np.random.uniform(size=500) -0.5     
X2= np.random.uniform(size=500) -0.5
y = 1 * (X1**2 - X2**2 > 0)
```

1b.

```python
plt.scatter(X1, X2, c=y)
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

1c.

```python
X = pd.DataFrame(X1, columns=["X1"])
X["X2"] = X2

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=200)

clf = LogisticRegression(random_state=0).fit(X, y)
```

1d.

```python
ypred = clf.predict(X)
plt.scatter(X1, X2, c=ypred)
plt.title("Predicted Y")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

1e.

```python
X_modified = X.copy()
X_modified['X1'] = X['X1']*X['X2']
X_modified['X2'] = X['X2']**2
clf = LogisticRegression(random_state=0).fit(X_modified, y)
```

1f.

```python
ypred = clf.predict(X_modified)
plt.scatter(X1, X2, c=ypred)
plt.title("Predicted Y")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

1g.

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X, y)
ypred = clf.predict(X)
plt.scatter(X1, X2, c=ypred)
plt.title("Predicted Y")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

The SVM classifier outperformed the logistic classifier with raw values of X1 and X2, and with non-linear functions of X1 and X2. The SVM classifier almost exactly captured the decision boundry we see in the raw data.
