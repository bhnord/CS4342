```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
X1 = [3, 2, 4, 1, 2, 4, 4]
X2 = [4, 2, 4, 4, 1, 3, 1]
y = [0, 0, 0, 0, 1, 1, 1]
col = ['blue', 'red']
plt.scatter(X1, X2, c = y, cmap='bwr_r')

X=np.linspace(1, 4.5, 50)
yplt = X-.5
plt.plot(X, yplt)
plt.fill_between(X, yplt, 5, color='red', alpha=.1)
plt.fill_between(X, yplt, -1, color='blue', alpha=.1)
plt.xlabel('X1')
plt.ylabel('X2')
```

```python

plt.scatter(X1, X2, c=y, cmap='bwr_r')
yplt1 = yplt - 0.5
yplt2 = yplt + 0.5
plt.plot(X, yplt)
plt.plot(X, yplt1, 'k--', alpha = 0.7)
plt.plot(X, yplt2, 'k--', alpha = 0.7)
plt.fill_between(X, yplt, 5, color='red', alpha=.1)
plt.fill_between(X, yplt, -1, color='blue', alpha=.1)
plt.xlabel('X1')
plt.ylabel('X2')
```

```python
plt.scatter(X1, X2, c=y, cmap='bwr_r')
yplt +=0.25
yplt1 = yplt - 0.25
yplt2 = yplt + 0.25
plt.plot(X, yplt)
plt.plot(X, yplt1, 'k--', alpha = 0.7)
plt.plot(X, yplt2, 'k--', alpha = 0.7)
plt.fill_between(X, yplt, 5, color='red', alpha=.1)
plt.fill_between(X, yplt, -1, color='blue', alpha=.1)
plt.xlabel('X1')
plt.ylabel('X2')
```
